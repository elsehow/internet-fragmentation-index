```python
import pandas as pd
from os import listdir
from os.path import join
from scipy.spatial.distance import cosine
import hashlib
```

```python
readings = listdir('data/country')
```

```python
ex = readings[0]
df = pd.read_csv(join('data/country', ex))
```

```python
originlist = lambda country_code: df[df['country_code']==country_code].sort_values(by='rank', ascending=True)['origin'].tolist()
originlist('us')
```

```python
cosine(originlist('us'), originlist('ca'))
```
